# **Questão 3**

```python
# Realizar um merge dos DataFrames base_carteira e base_originacao
base_carteira_merge = base_carteira.merge(base_originacao[['CPF', 'CONTRATO', 'QTPARCELAS']],
                                        on=['CPF', 'CONTRATO'], how='left')

# Criar uma lista vazia para armazenar os DataFrames de cada data base
data_frames_consolidados = []

# Agrupar o DataFrame 'base_carteira' por data base (DTBASE)
grouped_carteira = base_carteira_merge.groupby('DTBASE')

# Loop para processar cada data base e calcular as informações consolidadas
for data_base, grupo in grouped_carteira:
    # Quantidade de cooperativas distintas
    qtd_cooperativas = grupo['COOPERATIVA'].nunique()

    # Quantidade de CPFs distintos
    qtd_cpfs = grupo['CPF'].nunique()

    # Soma do saldo devedor
    saldo_devedor_total = round(grupo['SDDEVEDOR'].sum(), 1)

    # Média dos dias de atraso
    media_dias_atraso = round(grupo['QTDIASATRASO'].mean(), 1)

    # Menor data de originação dos contratos
    min_dtcontrato = pd.to_datetime(grupo['DTCONTRATO'], format='%d/%m/%Y').min()

    # Contrato com maior número de parcelas
    contrato_mais_parcelas = grupo.loc[grupo['QTPARCELAS'].idxmax(), 'CONTRATO']

    # Prazo médio ponderado dos contratos (valor da originação e quantidade de parcelas)
    prazo_medio_ponderado = round((grupo['VLCONTRATO'] * grupo['QTPARCELAS']).sum() / grupo['VLCONTRATO'].sum(), 2)

    # Taxa de contratos com atrasos > 90 dias
    taxa_atraso_90dias_contratos = round((grupo['QTDIASATRASO'] > 90).mean(),2)

    # Taxa de saldo devedor com atrasos > 90 dias
    taxa_atraso_90dias_saldo = round((grupo.loc[grupo['QTDIASATRASO'] > 90, 'SDDEVEDOR'].sum()) / saldo_devedor_total, 2)

    # Média da quantidade de parcelas
    media_qtd_parcelas = round(grupo['QTPARCELAS'].mean(), 2)

    # Criar um DataFrame com as informações da data base
    data_frame_consolidado = pd.DataFrame({'DTBASE': [data_base],
                                           'QTD_COOPERATIVAS': [qtd_cooperativas],
                                           'QTD_CPFS': [qtd_cpfs],
                                           'SALDO_DEVEDOR': [saldo_devedor_total],
                                           'MEDIA_DIAS_ATRASO': [media_dias_atraso],
                                           'MEDIA_QTD_PARCELAS': [media_qtd_parcelas],
                                           'MIN_DTCONTRATO': [min_dtcontrato],
                                           'CONTRATO_MAIS_PARCELAS': [contrato_mais_parcelas],
                                           'PRAZO_MEDIO_PONDERADO': [prazo_medio_ponderado],
                                           'TAXA_ATRASO_90DIAS_CONTRATOS': [taxa_atraso_90dias_contratos],
                                           'TAXA_ATRASO_90DIAS_SALDO': [taxa_atraso_90dias_saldo]})

    # Adicionar o DataFrame da data base à lista de DataFrames consolidados
    data_frames_consolidados.append(data_frame_consolidado)

# Concatenar todos os DataFrames consolidados em uma única tabela
tabela_consolidada = pd.concat(data_frames_consolidados, ignore_index=True)

# Ordenar a tabela consolidada pela DTBASE em ordem crescente
tabela_consolidada = tabela_consolidada.sort_values(by='DTBASE')

# Exibir a tabela consolidada
display(tabela_consolidada)
```

# **Questão 4**

Nessa questão vamos ter uma visão mensal (DTBASE), para o atributo QTPARCELAS, da Moda, e para o atributo SDDEVEDOR a Média e o Intervalo de Confiança para a média com nível de 99%.

O intervalo de confiança é uma medida estatística importante que nos ajuda a quantificar a incerteza em nossas estimativas de parâmetros populacionais com base em amostras. É um intervalo no qual temos um certo nível de confiança de que o verdadeiro valor do parâmetro está contido. Um nível de confiança de 99% significa que, em média, se repetirmos a amostragem e a análise muitas vezes, em cerca de 99% dos casos, o intervalo conterá o verdadeiro valor do parâmetro populacional. Ele vai nos fornecer uma medida de quão precisas são essas inferências, considerando a variabilidade dos dados amostrais.

```python
# Realizar o merge dos DataFrames base_carteira e base_originacao
base_carteira_merge = base_carteira.merge(base_originacao[['CPF', 'CONTRATO', 'QTPARCELAS']],
                                    on=['CPF', 'CONTRATO'], how='left')


# Criar uma lista vazia para armazenar os DataFrames de cada mês
data_frames_consolidados = []

# Agrupar o DataFrame 'base_carteira' por data base (DTBASE)
grouped_carteira = base_carteira_merge.groupby('DTBASE')

# Loop para processar cada mês e calcular as informações consolidadas
for data_base, grupo in grouped_carteira:
    # Calcular a Moda para a coluna QTPARCELAS
    moda_qtd_parcelas = grupo['QTPARCELAS'].mode().values[0]

    # Calcular a Média e o Intervalo de Confiança a 99% para a coluna SDDEVEDOR
    media_sddevedor = round(grupo['SDDEVEDOR'].astype(float).mean(),2)
    intervalo_confianca_sddevedor = stats.t.interval(0.99, len(grupo['SDDEVEDOR']) - 1, loc=media_sddevedor, scale=stats.sem(grupo['SDDEVEDOR'].astype(float)))

    # Criar um DataFrame com as informações do mês
    data_frame_consolidado = pd.DataFrame({'DTBASE': [data_base],
                                           'MODA_QTD_PARCELAS': [moda_qtd_parcelas],
                                           'MEDIA_SDDEVEDOR': [media_sddevedor],
                                           'IC_99_INFERIOR_SDDEVEDOR': [round(intervalo_confianca_sddevedor[0], 2)],
                                           'IC_99_SUPERIOR_SDDEVEDOR': [round(intervalo_confianca_sddevedor[1], 2)]})

    # Adicionar o DataFrame do mês à lista de DataFrames consolidados
    data_frames_consolidados.append(data_frame_consolidado)

# Concatenar todos os DataFrames consolidados em uma única tabela
tabela_consolidada = pd.concat(data_frames_consolidados, ignore_index=True)

# Exibir a tabela consolidada
display(tabela_consolidada)
```

Ao analisar os resultados apresentados na tabela, podemos identificar algumas tendências e informações importantes sobre os dados consolidados ao longo do ano de 2021:

**Tendências no Saldo Devedor Médio:** Podemos observar que o saldo devedor médio apresentou variações ao longo do ano. Iniciando em janeiro com um valor de aproximadamente 3965.10 e alcançando um pico em julho, com um valor próximo a 4180.99. A partir de então, houve uma leve queda em agosto e setembro, seguida de um aumento em outubro e novembro, e uma leve queda em dezembro, chegando a um valor próximo a 3875.36.

**Estabilidade na Moda da Quantidade de Parcelas:** A coluna "MODA_QTD_PARCELAS" mostra que a moda da quantidade de parcelas dos contratos é constante, permanecendo em 48 parcelas em todos os meses.

**Importância do Intervalo de Confiança:** O intervalo de confiança é essencial para avaliar a precisão das estimativas. Quanto mais estreito o intervalo de confiança, mais precisa é a estimativa da média. Nesse contexto, um intervalo de confiança mais amplo indica uma maior incerteza e variabilidade nos dados observados.

### **SQL (Problema 1-a)**

```python
import sqlite3
```

```python
# Criar uma conexão com o banco de dados
conn = sqlite3.connect('bancodedados.db')

# Inserir os dados do DataFrame na tabela 'base_originacao' no banco de dados
base_originacao.to_sql('base_originacao', conn, if_exists='replace', index=False)
```

```python
# Consulta SQL para calcular a quantidade de dias distintos em 'DTCONTRATO'
query_quantidade_dias_distintos = """
SELECT COUNT(DISTINCT DTCONTRATO) AS quantidade_dias_distintos
FROM base_originacao;
"""

# Consulta SQL para calcular a soma dos 'VLCONTRATO' dos contratos originados
query_soma_vlcontrato = """
SELECT SUM(VLCONTRATO) AS soma_vlcontrato
FROM base_originacao;
"""

# Executar as consultas SQL e armazenar os resultados em DataFrames
df_quantidade_dias = pd.read_sql_query(query_quantidade_dias_distintos, conn)
df_soma_vlcontrato = pd.read_sql_query(query_soma_vlcontrato, conn)

# Fechar a conexão com o banco de dados para otimizar a velocidade
conn.close()

# Exibir os resultados
quantidade_dias_distintos = df_quantidade_dias['quantidade_dias_distintos'][0]
soma_vlcontrato = df_soma_vlcontrato['soma_vlcontrato'][0]

print(f"Quantidade de dias distintos em DTCONTRATO: {quantidade_dias_distintos}")
print(f"Soma dos VLCONTRATO dos contratos originados: {soma_vlcontrato:.2f}")
```

### **PySparks (Problema 1-a)**

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, translate, countDistinct
from pyspark.sql import functions as F
```

```python
# Inicializar a sessão do Spark
spark = SparkSession.builder.appName("Exemplo Spark").getOrCreate()

# Carregar o arquivo CSV em um DataFrame, especificando o separador como ';'
base_originacao_df = spark.createDataFrame(base_originacao)


# Calcular a quantidade de dias distintos em 'DTCONTRATO'
quantidade_dias_distintos = base_originacao_df.select(countDistinct("DTCONTRATO")).collect()[0][0]

# Calcular a soma dos 'VLCONTRATO' dos contratos originados
soma_vlcontrato = base_originacao_df.select(F.sum("VLCONTRATO")).collect()[0][0]

# Exibir os resultados
print(f"Quantidade de dias distintos em DTCONTRATO: {quantidade_dias_distintos}")
print(f"Soma dos VLCONTRATO dos contratos originados: {soma_vlcontrato:.2f}")
```
